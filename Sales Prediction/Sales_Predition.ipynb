# Data Visualization

```python
# Visualization of TV against Sales
plt.figure(figsize=(8,6))
plt.plot(sale_df.TV,sale_df.Sales, '.')
plt.title("Relationship Between TV and Sales")
plt.xlabel("TV")
plt.ylabel("Sales")
plt.show()
```

```python
# Visualization of Radio against Sales

plt.figure(figsize = (8,6))
plt.plot(sale_df.Radio, sale_df.Sales, '.', color ='red')
plt.title("Relationship between Radio and Sales")
plt.xlabel("Radio")
plt.ylabel("Sales")
plt.show()
```

```python
# Finding the relationship that exist between Newspaper and Sales
plt.figure(figsize= (8,6))
plt.plot(sale_df.Newspaper,sale_df.Sales ,"." , color = "purple")
plt.title("Relationship between Newspaper and  Sales")
plt.xlabel("Newspaper")
plt.ylabel('Sales')
plt.show()
```

 ### Getting the visualizaton of the attributes in one visualization

```python
#combining all visualization
sns.pairplot(sale_df, y_vars=['Sales'], x_vars=["TV","Radio","Newspaper"])
```

```python
# Looking out for the correlation by value that exist in the dataset 

sale_corr = sale_df.corr()
sale_corr
```

```python
# Heatmap for the correlation that exist amongst the attributes

plt.figure(figsize=(8,6))
sns.heatmap(sale_corr, annot = True, cmap="Blues" )
```

### Building  Model (Linear Regression) for Prediction 

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split

X = sale_df[["TV","Radio",'Newspaper']]
y = sale_df["Sales"]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.4, random_state=100) #splitting the data into training and test model


reg = linear_model.LinearRegression()
reg.fit(X_train,y_train)
```

```python
reg.coef_
```

```python
print(reg.intercept_)
reg.score(X_test,y_test)
```
